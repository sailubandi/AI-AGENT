```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
powerful_llm = genai.GenerativeModel('gemini-1.5-flash-latest')
```

```python
from crewai import Agent
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata

# Initialize the Gemini API using the LangChain wrapper
powerful_llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash-latest",
    google_api_key=userdata.get("GOOGLE_API_KEY")
)

planner = Agent(
    role="Study Planner",
    goal="Design a 30-day smart study plan to crack a 40 LPA tech job",
    backstory="You're a professional academic planner. Your task is to structure realistic, targeted schedules.",
    allow_delegation=False,
    verbose=True,
    llm=powerful_llm
)

coach = Agent(
    role="Productivity Coach",
    goal="Add daily motivation and productivity tips",
    backstory="You're an expert in motivation and productivity, helping students stay focused and consistent.",
    allow_delegation=False,
    verbose=True,
    llm=powerful_llm
)

evaluator = Agent(
    role="Evaluator",
    goal="Review and refine the plan for balance and goal alignment",
    backstory="You're a senior advisor ensuring the study plan is effective and aligned with long-term goals.",
    allow_delegation=False,
    verbose=True,
    llm=powerful_llm
)
```

```python
from crewai import Task

plan_task = Task(
    description=(
        "Design a comprehensive 30-day study plan covering DSA, AI, ML, DL, Cybersecurity, and SCM. "
        "Ensure the user studies 3–4 hours/day outside college (9 AM–5 PM). Include rest and test days."
    ),
    expected_output="Day-wise schedule with time blocks, topics, and short goals.",
    agent=planner
)

coach_task = Task(
    description=(
        "Provide for each of the 30 days:\n"
        "- A motivational quote\n"
        "- A productivity or focus tip\n"
        "- Try aligning it with the subject planned that day"
    ),
    expected_output="30-day motivation and focus companion.",
    agent=coach
)

evaluate_task = Task(
    description=(
        "Evaluate and polish the 30-day study and motivation plan. "
        "Ensure it's realistic, aligned with the 40 LPA job goal, and avoids burnout."
    ),
    expected_output="Final markdown-formatted smart study planner.",
    agent=evaluator
)
```

```python
from crewai import Crew
from IPython.display import Markdown

crew = Crew(
    agents=[planner, coach, evaluator],
    tasks=[plan_task, coach_task, evaluate_task],
    verbose=2
)

result = crew.kickoff(inputs={"topic": "Cracking a 40 LPA job via Smart Study Plan"})
Markdown(result)
```
